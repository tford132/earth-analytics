# Spatial Data Formats Notes

## Raster Stored as Single Files (One file is one layer)

.tif/.tiff is the most common data format and stands for Tagged Image File Format

.asc Stands for **ASCII Raster Files**. This is a text based format that stores rast data. This format is used given its simple to store and distribute

## Hierarchical Data Formats

These files can store any different types of data in one single file. A bit more involved to work with by make processing more efficient.

.hdf / .hdf5 Stands for **Hierachical Data Format**. One of the most common hierachical was to store raster data. 

.nc (NetCDF) Stands for **Network Common Data Form**. A common way to store climate data.

It is probably worth checking how to use pyplot as it seems very important and used a lot especially to visualise rasters

```python
with rio.open(naip_pre_fire_url) as ref:
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (20,5))
    show((ref, 1), ax = ax1, cmap = "Reds")
    show((ref, 2), ax = ax2, cmap = "Greens")
    show((ref, 3), ax = ax3, cmap = "Blues")
    plt.show()
```

```python
with rio.open(naip_pre_fire_url) as ref:
    RGB = ref.read(2)
    show(RGB, transform = ref.transform)
```
